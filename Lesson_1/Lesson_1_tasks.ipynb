```python
import pandas as pd
```

```python
train_df = pd.read_csv('./data/train_tweets.csv', index_col='id')
test_df = pd.read_csv('./data/test_tweets.csv', index_col='id')
df = pd.concat((train_df, test_df))
df.head()
```

```python
df.info()
```

```python
import re
```

```python
def clean_text(text):
    # 1 Удалим @user из всех твитов
    text = re.sub('@\w*', ' ', text)
    # 2 Изменим регистр твитов
    text = text.lower() 
    # 3 Заменим сокращения с апострофами
    text = ' '.join(apostrophe_dict.get(key, key) for key in text.split())
    # 4 Заменим сокращения на их полные формы
    text = ' '.join(short_word_dict.get(key, key) for key in text.split())
    # 5 Заменим эмотиконы
    text = ' '.join(emoticon_dict.get(key, key) for key in text.split())
    # 6 Заменим пунктуацию на пробелы
    text = re.sub(r'[^\w\s]', ' ', text)
    # 7 Заменим спец. символы на пробелы
    text = re.sub(r'[^A-Za-z0-9]', ' ', text)
    # 8 Заменим числа на пробелы
    text = re.sub(r'[^A-Za-z]', ' ', text)
    # 9 Удалим из текста слова длиной в 1 символ
    text = ' '.join(word for word in text.split() if len(word) > 1)
    return text
```

```python
df['tweet'] = df['tweet'].apply(clean_text)
```

```python
# 10 Поделим твиты на токены
from nltk.tokenize import word_tokenize
df['tweet_token'] = df['tweet'].apply(word_tokenize)
```

```python
# 11 Удалим стоп-слова из токенов
from nltk.corpus import stopwords
def filter_tokens(tokens):
    return [token for token in tokens if token not in stopwords.words('english')]
```

```python
from multiprocessing import Pool
from tqdm import tqdm
with Pool(8) as p:
    df['tweet_token_filtered'] = tqdm(p.imap(filter_tokens, df['tweet_token']), total=len(df))
```

```python
# 12 Применим стемминг к токенам
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
def stem(text):
    return [stemmer.stem(word) for word in text]
```

```python
with Pool(8) as p:
    df['tweet_stemmed'] = tqdm(p.imap(stem, df['tweet_token_filtered']), total=len(df))
```

```python
# 13 Применим лемматизацию к токенам
from nltk.stem.wordnet import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatize(text):
    return [lemmatizer.lemmatize(word) for word in text]
```

```python
with Pool(8) as p:
    df['tweet_lemmatized'] = tqdm(p.imap(lemmatize, df['tweet_token_filtered']), total=len(df))
```
